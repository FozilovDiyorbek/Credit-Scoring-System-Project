```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv("C:/Projects/Credit_scoring_system_project/data/processed/cleaned_credit_rick.csv")
df.head()
```

```python
# Checking the Target column
print("Target distribution:\n", df["loan_status"].value_counts())
sns.countplot(x="loan_status", data=df)
plt.title("loan_status Distribution")
plt.show()
```

```python
# 4. Numerical columns
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
numeric_cols = [col for col in df.columns if df[col].dtype in ['float64', 'int64']]
numeric_cols.remove('loan_status')  

# Histograms
df[numeric_cols].hist(bins=30, figsize=(15, 10))
plt.suptitle("Numerical Features Distribution")
plt.tight_layout()
plt.show()
```

```python
corr = df[numeric_cols + ['loan_status']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
